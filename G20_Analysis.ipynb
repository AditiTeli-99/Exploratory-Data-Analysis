```python
def top_3(data, parameter):
    year_list = data["year"].dt.year.sort_values().unique()
    result_df = pd.DataFrame(columns=year_list,index=[1,2,3])
    
    # filling the above given dataframe with the help of for loop
    for year in year_list:
        df_year = data[data["year"].dt.year==year]
        value = df_year.groupby("country")[parameter].sum().sort_values(ascending=False).head(3).index
        result_df[year] = value
        
    return(result_df)
```

```python
for i in selected_cols:
    result = top_3(dff,i)
    print(i)
    display(result)
    print("---------------------------------------------------------")
```

## Life Expectancy

```python
life_expectancy = pd.pivot_table(columns="year_only",index="country",values="Life expectancy",data=dff,aggfunc="mean").round(1)
```

```python
life_expectancy
```

```python
## Percentage change in mean life expectancy:

life_expectancy["percentage_change"] = round((life_expectancy[2020] - life_expectancy[2015])*100/life_expectancy[2015],1)
```

```python
life_expectancy = life_expectancy.sort_values(ascending=False,by="percentage_change")
```

```python
life_expectancy
```

```python
# Let us visualize the above information through a bar graph:

plt.figure(figsize=(20,5))
sns.barplot(x=life_expectancy.index,y=life_expectancy["percentage_change"],palette="viridis")
plt.title("% Change in Life Expectancy")
plt.xticks(rotation=90)
plt.show()
```

## FDI

```python
fdi = pd.pivot_table(columns="year_only",index="country",values="Foreign direct investment (FDI)",data=dff,aggfunc="mean").round(1)
```
